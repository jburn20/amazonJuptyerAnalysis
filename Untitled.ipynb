```python
import pandas as pd 
df = pd.read_csv('anonAmazon.csv')
#read csv file to interact with amazon data
df = df.fillna(0) #fillna() replaces empty slots with argument

df["Total Charged"] = df["Total Charged"].str.replace('$','')
df["Total Charged"] = df["Total Charged"].astype(float)
#specify parts of dataframe with brackets, replace $ with nothing and change the type from str to float to do math


#df["Total Charged"].sum()
df["Total Charged"].max()
df["Total Charged"].min()
df["Total Charged"].median()
#math


```
